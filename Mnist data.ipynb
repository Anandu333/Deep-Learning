```python
# to raduce the dimension from 3d to 1d
# as because of 28*28 we get 3d therefore multiply it and reduce the dimension
X_train_flatten= X_train.reshape(len(X_train),28*28)
```

# Creating only one layer

```python
# first step create model
model= keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid'),
])
```

```python
# second step compile the model
model.compile(optimizer= 'adam',loss= "SparseCategoricalCrossentropy",metrics=['Accuracy'])
```

```python
# third step fit the model
model.fit(X_train_final,y_train,epochs=10)
```

```python
# convert X_test samples also to 3d to 2d
X_test_flatten= X_test.reshape(10000,784)
```

```python
X_test_final= X_test_flatten/255
```

```python
model.evaluate(X_test_flatten,y_test)
```

```python
predicted= model.predict(X_test_final)
```

# Creating two layer

```python
# while creating two layers no. of o/p required should be descending
# also pass i/p shape only once
model= keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])
```

```python
model.compile(optimizer= 'adam', loss='SparseCategoricalCrossentropy',metrics='Accuracy')
```

```python
# first time accuracy
model.fit(X_train_flatten,y_train)
```

```python
# accuracy with epochs
model.fit(X_train_flatten,y_train,epochs=10)
```

# Creating three layers

```python
#processing time increases as the layers increase
#layers increase accuracy increase and loss fnct decrease
model= keras.Sequential([
    keras.layers.Dense(1000, input_shape=(784,)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])
```

```python
model.compile(optimizer= 'adam', loss='SparseCategoricalCrossentropy',metrics='Accuracy')
```

```python
model.fit(X_train_flatten,y_train)
```

```python
model.fit(X_train_flatten,y_train, epochs=10)
```
